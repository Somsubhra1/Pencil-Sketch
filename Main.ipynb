```python
import cv2
from matplotlib import pyplot as plt
import math
```

```python
def displayImage(img, title=None, size=(10, 10)):
    plt.figure(figsize=size)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()
    
def displayImages(imgs, titles=None, size=(15, 15)):
    num = len(imgs)
    subplotnum = math.floor(math.sqrt(num)) + 1
    plt.figure(figsize=size)
    
    for idx, img in enumerate(imgs):
        ax = plt.subplot(subplotnum, subplotnum, idx+1)
        im = ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        if titles is not None:
            plt.title(titles[idx])
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
img = cv2.imread('./test_cases/1.jpg', 1)
displayImage(img, title="Original image", size=(5, 5))
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
displayImage(img_gray, title="Image in Gray Scale", size=(5, 5))
```

```python
img_invert = cv2.bitwise_not(img_gray)
displayImage(img_invert, title="Image Inverted", size=(5, 5))
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
displayImage(img_smoothing, title="Image Inverted", size=(5, 5))
```

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

```python
final_img = dodgeV2(img_gray, img_smoothing)
displayImage(final_img, title="Sketch", size=(7, 7))
```
